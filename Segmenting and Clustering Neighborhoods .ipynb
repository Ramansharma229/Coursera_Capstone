# I will be using the same approach which was present in the Clustering NewYork Lab Example.

```python
LIMIT = 100 # It will give 100 results which will be returned by the Foursquare API 
radius = 500
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

# Now By Using this function, I will create a dataframe which will give all the venues present in Toronto.

# Now I will Analyze each Neighbourhood and also use one-hot encoding.

```python
toronto_onehotencoding = pd.get_dummies(torontocan_venues[['Venue Category']], prefix="", prefix_sep="")
toronto_onehotencoding['Neighborhood'] = torontocan_venues['Neighborhood'] 
cols=list(toronto_onehotencoding.columns.values)
cols.pop(cols.index('Neighborhood'))
toronto_onehotencoding=toronto_onehotencoding[['Neighborhood']+cols]
toronto_onehotencoding.rename(columns = {'Neighborhood': 'Neighbourhood'}, inplace = True) # for consistent output 
toronto_onehotencoding.head()
```

```python
toronto_onehotencoding.shape
```

# Grouping the Neighbourhoods and taking their respective mean.

```python
torontocan_grouping = toronto_onehotencoding.groupby('Neighbourhood').mean().reset_index()
torontocan_grouping
```

# Let's get the top 5 common venues for each neighbourhood. 

```python
topcan_venues = 5

for hood in torontocan_grouping['Neighbourhood']:
    print("----"+hood+"----")
    temp = torontocan_grouping[torontocan_grouping['Neighbourhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(topcan_venues))
    print('\n')
```

# Let's Put all this information in the form of a dataframe.

```python
def common_venues(row, top_venues): # Sorting the venues in Descending Order 
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:top_venues]
                                            
top_venues = 10

indicators = ['st', 'nd', 'rd']
columns = ['Neighbourhood']
                                              
for ind in np.arange(top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))
neighborhoodscan_venues_sorted = pd.DataFrame(columns=columns)
neighborhoodscan_venues_sorted['Neighbourhood'] = torontocan_grouping['Neighbourhood']

for ind in np.arange(torontocan_grouping.shape[0]):
    neighborhoodscan_venues_sorted.iloc[ind, 1:] = common_venues(torontocan_grouping.iloc[ind, :], top_venues)

neighborhoodscan_venues_sorted
```

# Applying K-Means Clustering to create Clusters of the neighbourhoods.  

```python
kclusters = 5 # Defining the Number of Clusters

torontocan_grouping_kmeansclustering = torontocan_grouping.drop('Neighbourhood', 1)
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(torontocan_grouping_kmeansclustering)
kmeans.labels_[0:10]
```

# Let's Create a New DataFrame that combines the previous dataframe and the cluster as follows. 

```python
torontocan_merger = toronto_can 
torontocan_merger['Clustering Labels'] = kmeans.labels_ 
torontocan_merger = torontocan_merger.join(neighborhoodscan_venues_sorted.set_index('Neighbourhood'), on='Neighbourhood')
torontocan_merger.head()
```

# Let's Visualize the Clusters.

```python
torontocan_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
markers_colors = []
for lat, lon, poi, cluster in zip(torontocan_merger['Latitude'], torontocan_merger['Longitude'], torontocan_merger['Neighbourhood'], torontocan_merger['Clustering Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(torontocan_clusters)
       
torontocan_clusters
```
